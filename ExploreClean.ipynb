Read one of the voice command clips using scipy

```python
from google.colab import drive
drive.mount('/content/drive')
from pathlib import Path
example_audio_path = Path('/content/drive/MyDrive/Colab Notebooks/RecSpeech/ffd2ba2f_nohash_1.wav')
from scipy.io import wavfile
sample_rate, data = wavfile.read(example_audio_path)
```

```python
import numpy as np
pw = int(sample_rate * 0.01) #plot width
psi = 3000 #plot start index
print(sample_rate, data.shape, psi, pw)
```

```python
import matplotlib.pyplot as plt
plt.plot(data)
plt.show()
plt.plot(range(psi, psi+pw), data[psi : psi+pw], linestyle='None', marker='.',markersize=3)
```

Use librosa to load an audio clip from the dataset. Make sure the original sample rate is used. Then use the display functionality from librosa to plot a waveform with duration in seconds on the x-axis.

```python
import librosa
y, sr = librosa.load(example_audio_path, sr=None)
```

```python
sr
```

```python
fig, ax = plt.subplots(nrows=1, sharex=True)
ax.set(title='Envelope view')
librosa.display.waveshow(y, sr=sr, axis='time', ax=ax)
```

```python
fig, ax = plt.subplots(nrows=1, sharex=True)
ptw = 0.01 #plot time width
pst = 0.2 #plot start time
ax.set(xlim=[pst, pst+ptw], title='Sample view')
librosa.display.waveshow(y, sr=sr, axis='time', ax=ax, marker='.',linestyle='None')
```
